```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def gauss1d(data,mu,sigma):
    return (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-0.5*((data-mu)**2/sigma**2))
#def mu_simple(data):  #one unknown parameter (say, mu) need an array of sigmas
    #likeli_gauss1d(sig_list)
    #post=likeli*prior
```

```python
mu_t=5
sig_t=1
data=np.random.normal(mu_t,sig_t,3) #in the limit n tends to infinity the solution is ml
```

```python
#grid computing example
grid=100



std_prior_mu=0.8       #this prior represents what we think the unknown is, for example if sigma is unknown we might think sigma=1 (prior mu) and we are only about 1 sure (prior sigma) 
std_prior_sig=0.5

sig_u=np.linspace(0.1,4,grid)

prior= gauss1d(sig_u,std_prior_mu,std_prior_sig)   #prior where weighting (to likelihood values) is a 1d gaussian. This is essentially weighting the possible sig_u's (likelihoods)
#prior = np.repeat(1/grid, grid)  #this is a uniform prior, equal weighting to all sigs.
plt.plot(sig_u,prior)
plt.scatter(sig_u,prior,s=10)
plt.title('Prior sig unknown')
plt.show()
#print(sig_u)
likeli=np.zeros(len(sig_u))
likeli_test=np.zeros(len(sig_u))
for k,i in enumerate(sig_u):
    likeli[k]=np.prod(gauss1d(data,mu_t,i)) #likelihood of each sigma given the data.
    likeli_test[k] = norm(mu_t, i).pdf(data).prod()
plt.title('Likelihood of sigmas given data')
plt.plot(sig_u,likeli) #likeli hood of each sigma proposed given the data
plt.show()
post=likeli*prior   #not a proper pdf as the normalisation is not computed. In this case all the posterior can tell us is what the most likeli sigma is and it's std.
plt.plot(sig_u,post)
plt.title('Post sig unknown')
plt.scatter(sig_u,post,s=10)
```

```python
#metropolis algorithm
iters=2000
curr=0.8
delta=0.08
prior_mu=0.8
prior_std=0.3
trace=np.zeros(iters)
burn_in=50
for i in range(iters): #curr=current prop=proposed
    prop=np.random.normal(curr,delta)
    #print(prop)
    curr_likeli=np.prod(gauss1d(data,mu_t,curr)) #single value
    prop_likeli=np.prod(gauss1d(data,mu_t,prop))
    
    curr_prior=gauss1d(curr,prior_mu,prior_std) #single value
    prop_prior=gauss1d(prop,prior_mu,prior_std) #these two lines represent the prior probability of the curr and prop values 
    #print(curr_prior)
    curr_post=curr_likeli*curr_prior #the prior still acts as a weight. 
    #If the prior probability of the curr or prop values are low then it lowers the probability in posterior
    prop_post=prop_likeli*prop_prior 
    
    MH_crit=prop_post/curr_post  #metropolis hasting critreon ratio of the probabilities
    if MH_crit>np.random.rand():
        trace[i]=prop
        curr=prop
    else:
        trace[i]=curr
        
#print(trace[-1])
#plt.plot(trace)
plt.hist(trace[burn_in:])
print(np.mean(trace),np.std(trace))
```
