```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points=20
data1=np.linspace(0,10,points)
data2=data1+np.random.normal(0,1,points)
plt.scatter(data1,data2)
```

```python
#print(data1,data2)
```

```python
#this code shows that in RSS there is a true minimum value and not a continuum of minimums
#grid computing is done to calculate the residuals, not efficient.
def model(m,c,x):
    return np.multiply(m,x)+c
para_points=800
mspace=para_points
m=np.linspace(-2,2,mspace)
cspace=para_points
c=np.linspace(-2,2,cspace)
mm,cc=np.meshgrid(m,c)
z=np.zeros([mspace,cspace])
#print(z)
for k1,i in enumerate(m):
    for k2,j in enumerate(c):
        z[k1,k2]=np.sum((data2-model(i,j,data1))**2)    #the RSS part, have to set a m and c and calculate the residuals from of the observed from those.

fig,ax=plt.subplots()
im=ax.pcolormesh(c,m,z,cmap='viridis') #plot c and m, then plot their corresponding residual value
fig.colorbar(im)

r,col=np.where(z<np.min(z)+3) #find location of the residuals surrounding the minimum
cm = plt.cm.get_cmap('RdYlBu')
sc=plt.scatter(c[col],m[r],c=z[r,col],cmap=cm,s=1)  #plot residuals surrounding minimum
mmin,cmin=np.where(z==np.min(z))
print('min',np.min(z))
plt.colorbar(sc)
#ax.set_ylim([0.5,1.5])  #zoom in
#ax.set_xlim([-1.5,0.5])
plt.show()

print('y pos (m)',m[mmin],'x pos (c)',c[cmin])
print('yaxis',mmin,len(m),'xaxis',cmin,len(c))
#the fact that there is a ring of blue (large values) and a ring of red (small values) means there has to be a gradient that leads to a singular minimum.
```
